```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import STOPWORDS
```

```python
from wordcloud.wordcloud import WordCloud
```

```python
names = ['vamshikrishna vk', 'tharuni', 'sahana', 'venkatesh v&s', 'raviteja', 'tharun tharun_tamilnadu_alludu', 'sirichandhana siri', 'prudhvi', 'saiteja', 'vishnu gupta', 'soundarya sound this_is', 'mounika pichuka', 'lavanya', 'vyshnavi']
```

```python
heading = 'Friends forever besties btech cse-c kare klu batch trips canteen '*150

st = ''
st += heading
```

```python
# for better view run this cell 3 times
for i in names:
    times = np.random.randint(60,100)
    name = i + " "
    st += name * times 

print(st)
```

```python
wordart = WordCloud(width = 500, height = 400, stopwords = STOPWORDS, background_color ='white').generate(st) 
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wordart)
plt.axis("off") 
plt.show()
```
